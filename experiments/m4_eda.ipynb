```python
# ACF/PACF utilities for M4 samples

import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import pacf as sm_pacf

from src.data.preprocessing import tsf_series_load, uniform_resample, compute_paa_series, compute_paa_trend_segments, compute_paa_trend_directions

SERIES_INDEX = 0
PAA_SEGMENTS = 10

# How many series to use when estimating an aggregate lag for a whole set.
# Increase if you want a more stable estimate (slower).
LAG_SAMPLE_SIZE = 200

# Upper bound for PACF lag search on raw/interpolated series.
MAX_LAG_RAW_INTERP = 60


def _to_1d_float(series):
    arr = np.asarray(series, dtype=float).reshape(-1)
    arr = arr[np.isfinite(arr)]
    return arr


def most_common_length(series_list):
    lengths = [len(s) for s in series_list]
    return Counter(lengths).most_common(1)[0][0]


def _safe_lags(n: int, preferred: int):
    # statsmodels requires lags < n; PACF is typically limited to < n/2.
    if n < 3:
        return 0
    return int(max(1, min(preferred, (n // 2) - 1, n - 1)))


def _plot_acf_pacf(series, title: str, preferred_lags: int = 40):
    series = _to_1d_float(series)
    lags = _safe_lags(len(series), preferred_lags)
    if lags <= 0:
        print(f"{title}: not enough data points for ACF/PACF (n={len(series)})")
        return

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    plot_acf(series, lags=lags, ax=axes[0], title=f"ACF — {title}")
    plot_pacf(series, lags=lags, ax=axes[1], title=f"PACF — {title}", method="ywm")
    plt.tight_layout()
    plt.show()


def prepare_series_quad(tsf_series_list, series_index: int, target_len: int, paa_segments: int):
    raw = _to_1d_float(tsf_series_list[series_index])
    interpolated = uniform_resample(raw, target_len)
    paa = np.asarray(compute_paa_series(interpolated, paa_segments), dtype=float)
    trend_deltas, trend_segments = compute_paa_trend_segments(interpolated, paa_segments)
    trend_deltas = np.asarray(trend_deltas, dtype=float)
    trend_dirs = compute_paa_trend_directions(interpolated, paa_segments, threshold=0.0)
    return raw, interpolated, paa, trend_deltas, trend_segments, trend_dirs


def plot_quad(period_label: str, raw, interpolated, paa):
    _plot_acf_pacf(raw, f"{period_label} (raw)", preferred_lags=40)
    _plot_acf_pacf(
        interpolated,
        f"{period_label} (interpolated, len={len(interpolated)})",
        preferred_lags=40,
    )
    _plot_acf_pacf(paa, f"{period_label} (PAA, segments={len(paa)})", preferred_lags=10)



def estimate_series_lag_pacf(series, max_lag: int) -> int:
    """Estimate a single lag for one series using the PACF cutoff heuristic.

    Definition: the largest lag whose PACF is significantly different from 0 (95% CI).
    Returns 0 when no lag is significant or when not enough data.
    """
    series = _to_1d_float(series)
    lags = _safe_lags(len(series), max_lag)
    if lags <= 0:
        return 0

    values, confint = sm_pacf(series, nlags=lags, method="ywm", alpha=0.05)

    sig = []
    for k in range(1, len(values)):
        lo, hi = confint[k]
        if lo > 0 or hi < 0:
            sig.append(k)

    return max(sig) if sig else 0


def aggregate_lag_pacf(series_list, *, max_lag: int, sample_size: int, seed: int = 42):
    """Per-set aggregate lag = median of per-series PACF cutoff lags."""
    rng = np.random.default_rng(seed)
    n = len(series_list)
    if n == 0:
        raise ValueError("series_list is empty")

    if sample_size <= 0 or sample_size >= n:
        indices = np.arange(n)
    else:
        indices = rng.choice(n, size=sample_size, replace=False)

    per_series = [estimate_series_lag_pacf(series_list[i], max_lag=max_lag) for i in indices]
    per_series_sorted = sorted(per_series)

    median = int(np.median(per_series_sorted))
    mode = int(Counter(per_series_sorted).most_common(1)[0][0])

    return {
        "n_total": int(n),
        "n_used": int(len(indices)),
        "median": median,
        "mode": mode,
        "min": int(per_series_sorted[0]),
        "max": int(per_series_sorted[-1]),
        "top5": Counter(per_series_sorted).most_common(5),
    }
```

```python
# Hourly: ACF/PACF for raw vs interpolated vs PAA

tsf_path = root / "src" / "data" / "datasets" / "M4" / "m4_hourly_dataset.tsf"
series_list = tsf_series_load(tsf_path)

# Use the most common raw series length in this set to minimise information loss.
TARGET_LEN = most_common_length(series_list)

# Aggregate 'optimal lag' for this set using PACF cutoff per-series, then median across series.
raw_lag = aggregate_lag_pacf(
    series_list,
    max_lag=MAX_LAG_RAW_INTERP,
    sample_size=LAG_SAMPLE_SIZE,
)

interpolated_series_list = [uniform_resample(s, TARGET_LEN) for s in series_list]
interp_lag = aggregate_lag_pacf(
    interpolated_series_list,
    max_lag=MAX_LAG_RAW_INTERP,
    sample_size=LAG_SAMPLE_SIZE,
)

paa_series_list = [compute_paa_series(s, PAA_SEGMENTS) for s in interpolated_series_list]
paa_lag = aggregate_lag_pacf(
    paa_series_list,
    max_lag=PAA_SEGMENTS - 1,
    sample_size=LAG_SAMPLE_SIZE,
)




# Sanity check: show one PAA example before plotting.
raw, interpolated, paa, trend_deltas, trend_segments, trend_dirs = prepare_series_quad(
    series_list,
    series_index=SERIES_INDEX,
    target_len=TARGET_LEN,
    paa_segments=PAA_SEGMENTS,

)
trend_dirs_list = [compute_paa_trend_directions(s, PAA_SEGMENTS, threshold=np.std(trend_deltas)) for s in interpolated_series_list]
trend_dirs_counts = Counter(int(v) for dirs in trend_dirs_list for v in dirs)

print(f"Hourly: TARGET_LEN={TARGET_LEN}, PAA_SEGMENTS={PAA_SEGMENTS}, SAMPLE={LAG_SAMPLE_SIZE}")
print("Aggregate lag (raw):", raw_lag)
print("Aggregate lag (interpolated):", interp_lag)
print("Aggregate lag (PAA):", paa_lag)
print("Trend direction counts (-1/0/1):", dict(trend_dirs_counts))
print("Example PAA:", paa.tolist())
print("Example trend endpoints (first,last):", [(s["first"], s["last"]) for s in trend_segments])
print("Example trend deltas:", trend_deltas.tolist())
print("Standard deviation of trend deltas:", np.std(trend_deltas))
print("Example trend directions (-1/0/1):", trend_dirs_list[0])

plot_quad("Hourly", raw, interpolated, paa)
```
