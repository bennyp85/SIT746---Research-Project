```python
# ============================================
# Correct trend-threshold orchestration logic
# ============================================

tsf_path = Path("data/M4") / "m4_hourly_dataset.tsf"
series_list = tsf_series_load(tsf_path)

# Use the most common raw series length to minimise information loss
TARGET_LEN = most_common_length(series_list)

# Interpolate all series to a common length
interpolated_series_list = [
    uniform_resample(s, TARGET_LEN) for s in series_list
]

# Compute PAA values (used later for quantum encoding)
PAA_SEGMENTS = 10
paa_series_list = [
    compute_paa_series(s, PAA_SEGMENTS) for s in interpolated_series_list
]

# ------------------------------------------------
# Trend direction computation (PER SERIES)
# ------------------------------------------------

K_MULTIPLIER = 1.0  # sensible default for M4 (daily/weekly/hourly)

trend_dirs_list = []

for s in interpolated_series_list:
    # 1. Compute raw intra-segment trend deltas
    #    (this must match the logic used inside paa_trend_directions)
    raw_trend_deltas = compute_raw_trend_deltas(
        series=s,
        window_size=len(s) // PAA_SEGMENTS
    )

    # 2. Compute per-series scale
    sigma_delta = np.std(raw_trend_deltas)

    # 3. Define adaptive threshold
    threshold = K_MULTIPLIER * sigma_delta

    # 4. Compute trend directions using a scale-consistent threshold
    trend_dirs = paa_trend_directions(
        series=s,
        window_size=len(s) // PAA_SEGMENTS,
        threshold=threshold
    )

    trend_dirs_list.append(trend_dirs)

# ------------------------------------------------
# Diagnostics (optional but recommended)
# ------------------------------------------------

trend_dirs_counts = Counter(
    int(v) for dirs in trend_dirs_list for v in dirs
)

print("Trend direction counts:", trend_dirs_counts)
```
