```python

from src.data.preprocessing import tsf_series_load
from src.data.preprocessing import uniform_resample
tsf_path = root / "src" / "data" / "datasets" / "M4" / "m4_daily_dataset.tsf"
series_list = tsf_series_load(tsf_path)
resampled_series_list = [uniform_resample(s, 100) for s in series_list]
print("Original length of first series:", len(series_list[0]))
print("Resampled length of first series:", len(resampled_series_list[0]))
print("First 10 values of first resampled series:", resampled_series_list[0][:10])
```

```python
# plot original and resampled series side by side
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(series_list[1])
axs[0].set_title("Original Series")
axs[1].plot(resampled_series_list[1])
axs[1].set_title("Resampled Series")
plt.show()
```

```python
from src.data.preprocessing import compute_paa_series
paa_series_list = [compute_paa_series(s, 10) for s in resampled_series_list]
print("PAA length of first series:", len(paa_series_list[1]))
print("PAA series:", paa_series_list[0])
```

```python
from src.data.preprocessing import paa_trend_directions
trend_directions = paa_trend_directions(resampled_series_list[1], 10, 5)
print("PAA trend directions of first series:", trend_directions)
```
